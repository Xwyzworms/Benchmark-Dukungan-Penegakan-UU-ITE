```python
def ReturnNonSame():
    indexLocation = []
    for iteration,i in enumerate(HateSpeechGroup["Tweet"]):
        if (i not in HateSpeechRace["Tweet"].values) and (i not in HateSpeechReligion["Tweet"].values) and (i not in HateSpeehPhysical["Tweet"]):
            indexLocation.append(iteration)
    return indexLocation
print(f"HateSpeech : {len(ReturnNonSame())}")
```
